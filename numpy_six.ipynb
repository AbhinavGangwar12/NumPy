```python
import matplotlib.pyplot as p
# matplotlib takes the size of the matrix or image in the form of (rows,columns,depth)
img = np.array(p.imread("test.webp"))
p.imshow(img)
```

```python
img.shape#this is basically a numpy array with depth 1127 rows 1600 and 3 colums
# since matplotlib has a different viewport so it reads this data as columns = 1127, rows = 1600, depth = 3
# now when we print this as numpy array you will observe that 1 row contains three columns only and single row of a matrix is considered as a pixel
img
```

```python
# NOw we can play with the values of the image and make the changes in actual image
p.imshow(img[::-1,:,:])#reversing rows
```

```python
p.imshow(img[:,::-1,:])#reversing columns
```

```python
p.imshow(img[:,:,::-1])#reversing depth
```

```python
p.imshow(img[600:1100:,600:1300,:])#slicing-->zooming
```

```python
p.imshow(img[600:1100:,600:1300,(0)])#slicing-->zooming and indexing the depth too
```

```python
# Blurring the image
p.imshow(img[::100,::100,:])
```

```python
# Blurring the image
p.imshow(img[::50,::100,:])
```

```python
# applying filter
img_new = np.where(img>140,240,0)
p.imshow(img_new)
```

SIMILARLY WE CAN DO MORE OPERATIONS ON THESE KIND OF THINGS

```python
# Blurring the image
new = img.copy()
new[:,:,(0,2)]=0
p.imshow(new)
```

```python
blue = img.copy()
blue[:,:,(1,2)]=0
p.imshow(blue)
```
